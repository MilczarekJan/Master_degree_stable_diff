```python
import os
import pydicom
from PIL import Image
import numpy as np

# Ścieżka do folderu z plikami DICOM
input_folder = r"G:\Studia\Magisterka\Data\tcga_luad\TCGA-17-Z011\03157\41262"

# Utworzenie folderu wyjściowego
output_folder = input_folder + "_png"
os.makedirs(output_folder, exist_ok=True)

# Przetwarzanie wszystkich plików .dcm
for filename in os.listdir(input_folder):
    if filename.lower().endswith(".dcm"):
        input_path = os.path.join(input_folder, filename)

        # Wczytanie DICOM
        ds = pydicom.dcmread(input_path)
        img = ds.pixel_array.astype(np.float32)

        # Skalowanie do 0-255
        img = (img - img.min()) / (img.max() - img.min()) * 255
        img = img.astype(np.uint8)

        # Nazwa pliku PNG
        output_name = os.path.splitext(filename)[0] + ".png"
        output_path = os.path.join(output_folder, output_name)

        # Zapis
        Image.fromarray(img).save(output_path)

        print(f"Zapisano: {output_name}")

print("Gotowe!")
```

```python
import os
from PIL import Image

# ===== USTAWIENIA =====

input_folder = r"G:\Studia\Magisterka\Data\tcga_luad\TCGA-17-Z011\03157\41262_png"

# ======================

# Tworzenie folderu wyjściowego obok wejściowego
parent_folder = os.path.dirname(input_folder)
folder_name = os.path.basename(input_folder)

output_folder = os.path.join(
    parent_folder,
    folder_name + "_separated"
)

os.makedirs(output_folder, exist_ok=True)


# Obsługiwane formaty
extensions = (".jpg", ".jpeg", ".png", ".webp", ".bmp")


for filename in os.listdir(input_folder):

    if not filename.lower().endswith(extensions):
        continue

    input_path = os.path.join(input_folder, filename)

    try:
        img = Image.open(input_path)

        width, height = img.size

        # Punkty podziału
        mid_x = width // 2
        mid_y = height // 2

        parts = [
            ("_1", (0, 0, mid_x, mid_y)),             # lewy góra
            ("_2", (mid_x, 0, width, mid_y)),         # prawy góra
            ("_3", (0, mid_y, mid_x, height)),        # lewy dół
            ("_4", (mid_x, mid_y, width, height))    # prawy dół
        ]

        name, ext = os.path.splitext(filename)

        for suffix, box in parts:
            cropped = img.crop(box)

            output_name = f"{name}{suffix}{ext}"
            output_path = os.path.join(output_folder, output_name)

            cropped.save(output_path)

        print(f"OK: {filename}")

    except Exception as e:
        print(f"Błąd przy {filename}: {e}")


print("\nGotowe!")
print("Zapisano do:", output_folder)
```
